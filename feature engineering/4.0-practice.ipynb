## Minimum / maxium / median / Q1,Q3,IQR


```python
import numpy as np
```

```python
lst_marks= [45,32,56,75,89,26,67,88,93,87,84,67,74,99,96]
minimum,Q1,median,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.5,0.75,1.0])
```

```python
print(minimum,Q1,median,Q3,maximum)
```

```python
IQR = Q3-Q1
```

```python
IQR
```

```python
lower_fence = Q1 - 1.5*(IQR)
upper_fence = Q3 + 1.5*(IQR)
```

```python
lower_fence,upper_fence
```
